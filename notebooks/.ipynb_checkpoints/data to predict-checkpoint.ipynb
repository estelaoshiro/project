```python
import pandas as pd
```

```python
data = pd.read_csv("../data/df_selection_predict.csv", index_col=0)
```

```python
data
```

```python
# First, you need to convert the DATE and TIME columns to datetime format
data['DATE_TIME'] = pd.to_datetime(data['DATE'] + ' ' + data['TIME'])

# Group by hour
calls_per_hour = data.groupby(pd.Grouper(key='DATE_TIME', freq='H'))['UNIQUE_ID'].count()



```

```python
calls_per_hour
```

```python
data_filter = pd.DataFrame({'DATE_TIME':calls_per_hour.index, 'calls_per_hour':calls_per_hour.values})
data_filter
```

```python
start_date = '2017-01-01'
end_date = '2017-01-08'
mask = (data_filter['DATE_TIME'] >= start_date) & (data_filter['DATE_TIME'] <= end_date)
filter_data = data_filter.loc[mask].reset_index(drop=True)
filter_data
```

```python
filter_data = filter_data.iloc[:168,:]
```

```python
filter_data.to_csv("data_to_predict.csv")
```

```python
data = data[data['Open Data Channel Type']!="UNKNOWN"]
```

```python
filter2=data.groupby(['Complaint Type'])['Unique Key'].count().sort_values(ascending=False)
```

```python
filter2_df = pd.DataFrame({'Complaint Type': filter2.index, 'Count': filter2.values})
top_30 = filter2_df['Complaint Type'].head(30)
```

```python
data = data[data['Complaint Type'].isin(top_30)]
```

```python
data.shape
```

```python
data.dtypes
```

```python
data['diff'] = pd.to_timedelta(data['diff'])
```

```python

data['diff'] = data['diff'].dt.days
```

```python
data.head()
```

```python
data = data[data['diff']<=80]
```

```python
data = data[data['Open Data Channel Type'] != "OTHER"]
```

```python
data.shape
```

```python
data.groupby(['City'])['Unique Key'].count().sort_values(ascending=True)
```

```python
def upper_case(x):
    return x.upper()
```

```python
data['City'] = data['City'].apply(upper_case)
```

```python
city_counts = data.groupby(['City'])['Unique Key'].count().sort_values(ascending=True)
city_counts_df = pd.DataFrame({'City': city_counts.index, 'Count': city_counts.values})
```

```python
city_counts_df.tail(42)
```

```python
top_city = city_counts_df['City'].tail(42)
```

```python
data = data[data['City'].isin(top_city)]

```

```python
data.shape
```

```python
data.head()
```

```python
data = data.drop(['Closed Date'], axis=1)
data
```

```python
data = data[data['Open Data Channel Type']=='PHONE']
```

```python
import pandas as pd
import geopandas as gpd
```

```python
data = pd.read_csv('data_tableau2.csv', index_col=0)
```

```python

geometry = gpd.points_from_xy(data.Longitude, data.Latitude)
gdf = gpd.GeoDataFrame(data, geometry=geometry)

# Load the New York City neighborhoods shapefile from a URL
url = 'https://data.cityofnewyork.us/api/geospatial/cpf4-rkhq?method=export&format=Shapefile'
nyc_neighborhoods = gpd.read_file(url)

# Spatially join the two GeoDataFrames to get the neighborhood names
joined = gpd.sjoin(gdf, nyc_neighborhoods, op='within')

# Add the neighborhood names to the original DataFrame
data['neighborhood'] = joined['ntaname']
```

```python
data
```

```python
filter_data = ['OVERFLOWING LITTER BASKETS', 'VIOLATION OF PARK RULES', 'TRAFFIC']
```

```python
mask = ~data['Complaint Type'].isin(filter_data)
data = data.loc[mask]
```

```python
data = data.drop(['City'], axis=1)
```

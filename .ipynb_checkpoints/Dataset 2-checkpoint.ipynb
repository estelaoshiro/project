```python
file_path = '311_Service_Requests_from_2010_to_Present.csv'
chunk_size = 1000000

dataframes = []

for chunk in pd.read_csv(file_path, usecols=['Unique Key', 'Created Date', 'Closed Date', 'Agency', 'Agency Name',       'Complaint Type', 'Descriptor', 'Location Type', 'Incident Zip',       'City', 'Due Date', 'X Coordinate (State Plane)',       'Y Coordinate (State Plane)', 'Open Data Channel Type',       'Park Borough',       'Latitude', 'Longitude', 'Location'], chunksize=chunk_size):
    chunk = chunk[(chunk['Created Date']  > '01/01/2012') & (chunk['Created Date'] < '31/12/2019') & (chunk['Agency'] == '3-1-1')][['Unique Key', 'Created Date', 'Closed Date', 'Agency', 
                                                                                                    'Agency Name',
       'Complaint Type', 'Descriptor', 'Location Type', 'Incident Zip',
       'City', 'Due Date', 'X Coordinate (State Plane)',
       'Y Coordinate (State Plane)', 'Open Data Channel Type',
       'Park Borough',
       'Latitude', 'Longitude', 'Location']]
    dataframes.append(chunk)
    
df = pd.concat(dataframes)

print(df.head())
```
